```python
from time import time
```

```python
while True:
  try:
    t_start = time()
  
    df = next(df_iter)

    df.tpep_pickup_datetime = pd.to_datetime(df.tpep_pickup_datetime)  # from TEXT to TIMESTAMP
    df.tpep_dropoff_datetime = pd.to_datetime(df.tpep_dropoff_datetime)  # from TEXT to TIMESTAMP

    df.to_sql(name='yellow_taxi_data', con=engine, if_exists='append')
  
    t_end = time()

    print('Inserted another chunk..., took %.3f second' % (t_end - t_start))
  except StopIteration:
    print('Finished iterating through all data chunks.')
    break
```
